```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/Users/dipeshkumar/Downloads/titanic.csv')
```

```python
data
```

```python
data.drop(['name','sibsp','parch','ticket','cabin','embarked','boat','body','home.dest'],axis='columns',inplace=True)
```

```python
data
```

```python
target = data.survived
inputs = data.drop(['survived'],axis='columns')
```

```python
dummies = pd.get_dummies(inputs.sex)
```

```python
dummies.head()
```

```python
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head()
```

```python
inputs.drop('sex',axis='columns')
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.age = inputs.age.fillna(inputs.age.mean())
```

```python
inputs
```

```python
from sklearn.model_selection import train_test_split
x_train,y_train,x_test,y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
len(x_train)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test)
```

```python
model.predict(x_test)
```
